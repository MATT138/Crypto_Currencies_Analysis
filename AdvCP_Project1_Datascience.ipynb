#1st Question

Which coins grow the most in the past few years?

```python
#For question 1
average_df = pd.DataFrame()
i=0
for e in dfs:
  e['Date']=pd.to_datetime(e['Date'],yearfirst=True)
  e.set_index('Date',inplace=True)
  average_df[dfs_name[i]]=(e['Close']+e['Open'])/2
  i+=1
average_df
```

##Graph 1.1

```python
from matplotlib.pyplot import figure
colors=['red','blue','green','yellow','orange','pink','cyan','purple','brown','black','grey']
average_df.reset_index(inplace=True)
sb.set(rc={'figure.figsize':(13,7)})
for i in range(11):
  ax=sb.lineplot(x='Date',y=dfs_name[i],data=average_df,color=colors[i],label=dfs_name[i])
ax.set(xlabel='Date',ylabel='value')
plt.show()
```

From this graph, it can be observed that BTC has grown by the largest amount over this 5-6 years follow by ETH and BNB. However, each coins start at different value. So it might be unfair for cheaper coins.
*For example
A 100->150 grow by 1.5 times, value increase by 50
B 1->1.5 also grow by 1.5 times, however value only increase by 0.5

```python
colors_df=pd.DataFrame()
colors_df['Currencies']=dfs_name
colors_df['colors']=colors
colors_df=colors_df.set_index('Currencies')
colors_df
```

```python
growth_df=pd.DataFrame()
growth_df['Date']=average_df['Date']
start = []
for e in dfs:
  l=list(e['Open'])
  for eil in l:
    if not pd.isna(eil):
      start.append(eil)
      break
i=0
for e in dfs_name:
  growth_df[e]=average_df[e]/start[i]
  i+=1
growth_df
```

**When is the best time to invest?**

**When is the best time to invest for ETH?**

```python
import datetime
monthLow_df=pd.DataFrame()
monthLow_df['Month']=[i for i in range(1,13)]
monthLow_df.set_index('Month',inplace=True)
for i in range(2015,2022):
  low_df=pd.DataFrame()
  low_df['Month']=ETH.reset_index()['Date'].dt.month
  low_df['Year']=ETH.reset_index()['Date'].dt.year
  low_df['ETH']=ETH.reset_index()['Low']
  yearLow=low_df.loc[low_df['Year']==i]
  yearLow=yearLow.groupby('Month').min()
  monthLow_df[i]=yearLow['ETH']
monthLow_df.reset_index(inplace=True)
monthLow_df
```

##Graph 3.1

```python
plt.clf()
for i in range(2015,2022):
  ax=sb.lineplot(x='Month',y=i,data=monthLow_df,color=colors[i-2015],label=i)
  ax.set(xlabel='Month',ylabel='Lowest of each month',title='Lowest value of ETH in each month')
plt.show()
```

##Conclude Graph 3.1

```python
import math
month_names=['January','February','March','April','May','June','July','August','September','October','November','December']
l=[]
for i in range(2015,2022):
  pair=[math.inf,-1]
  j=0
  for eil in list(monthLow_df[i]):
    if not pd.isna(eil) and eil<pair[0]:
      pair[0]=eil
      pair[1]=j
    j+=1
  l.append(month_names[pair[1]])
table=pd.DataFrame()
table['Year']=[i for i in range(2015,2022)]
table['Month with lowest Low']=l
table.set_index('Year',inplace=True)
table
```

For ETH, January seem to be the best month to invest. Since for 3 out of 7 years, this month has the lowest 'Low' compared to other months.
